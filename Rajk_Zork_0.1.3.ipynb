# Kiegészítő feature ötletek kurzushétvégére:
* mentés - betöltés opció
* tárgyak felvétele, elraktározás, későbbi felhasználás
* Lili féle tarot kártyahúzás integrálása
* statisztikák gyűjtése a játékről, tárolás -> összstatisztikák mutatása
* játék nehézség - könnyű (1 NBSZ), közepes (1 NBSZ 1 strigula), nehéz (1 NBSZ 2 strigula)
* verziókövetés - korábbi föntések hogyan befolyásolják a mostani döntési opciót
    * sima milyen döntési út miatt jutottál ide alapján (pl. azután turkálsz-e a zsebedben, hogy sörös lett a kezed)
    * skála alapján (pl. jó-rossz, szakmai példakép-antikrisztus skála)
    * nehézség alapján (pl. kizáródsz-e egy udvaron hirtelen kapott strigulától)
    * számszerűsíthető kummulált dolgok alapján (pl. hány cigit szívtál eddig el)
* minden döntésnek van egy idő költsége, ha kifutsz az időből az is lehet NBSz
* vizuális interface - koliban épp mi hol van (pl. Joza videójában szereplő 3d-s modell vagy a körbefotózott kollégiumi épület vagy ilyesmi alaopján)
